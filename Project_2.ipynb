```python
# Importing pandas library
import pandas as pd
```

```python
# Import the youtube dataset
df_youtube = pd.read_csv('youtube_dataset.csv',encoding = "iso-8859-1")
df_youtube
```

### euro symbol is represented as '\x80' in latin-1 encoding. so, we are using replace method to remove that euro symbol  

```python
# Replacing the euro symbol by empty space in MonthlyEarnings column
df_youtube.loc[:,'MonthlyEarnings'] = df_youtube['MonthlyEarnings'].str.replace('\x80','')
```

```python
# Replacing the euro symbol by empty space in YearlyEarnings column
df_youtube.loc[:, 'YearlyEarnings'] = df_youtube['YearlyEarnings'].str.replace('\x80', '')
```

```python
# Displaying the dataset after replacing Euro symbol
df_youtube
```

### MonthlyEarnings and YearlyEarnings column has a range value indicating the minimum and maximum earning separated by '-'. Also It contains characters K and M indicating earnings in thousands and millions. Here we are trying to split the Monthly earnings into Minimum and Maximum MonthlyEarnings. Then we will convert the Minimum and maximum MonthlyEarning into numeric value

```python
# Split the MonthlyEarnings column into Min and Max
df_youtube[['MinYearlyEarnings', 'MaxYearlyEarnings']] = df_youtube['YearlyEarnings'].str.split(' - ',expand = True)
```

```python
# Split the MonthlyEarnings column into Min and Max
df_youtube[['MinMonthlyEarnings', 'MaxMonthlyEarnings']] = df_youtube['MonthlyEarnings'].str.split(' - ',expand = True)
```

### We have created our own function convertEarnings to convert the Minimum and maximum earnings into numerical value. Suppose 66.9K earning will be converted as 66900 and 12.8M will be converted as 12800000. So, the min and max monthly earnings will be of type Float

```python
# Function to convert the min and max earnings into a numerical value 
def convertEarnings(value):
    # Checkes if charcter 'K' is present in earnings
    if 'K' in value:
        # Remove K and convert the earnings to float and multiply by 1000
        value = float(value.replace('K',''))*1000
    # Checkes if charcter 'K' is present in earnings
    elif 'M' in value:
        # Remove M and convert the earnings to float and multiply by 1000000
        value = float(value.replace('M',''))*1000000
    return value
```

### After defining our function convertEarnings, we need to apply this conversion to each value in Min and Max earnings column. so, we using apply method and passing the function as input which in turn calls the function and applies the conversion for each value in the column

```python
# Converting the Min and Max earnings to numerical value
df_youtube['MinMonthlyEarnings'] = df_youtube['MinMonthlyEarnings'].apply(convertEarnings)
df_youtube['MaxMonthlyEarnings'] = df_youtube['MaxMonthlyEarnings'].apply(convertEarnings)
df_youtube['MinYearlyEarnings'] = df_youtube['MinYearlyEarnings'].apply(convertEarnings)
df_youtube['MaxYearlyEarnings'] = df_youtube['MaxYearlyEarnings'].apply(convertEarnings)
```

```python
# Displaying the dataset after applying the conversion to numerical columns
df_youtube
```

### Since we have extracted the minimum and maximum earnings, we can drop the columns MonthlyEarnings and YearlyEarnings

```python
# Dropping the MonthlyEarnings and YearlyEarnings columns
df_youtube.drop(columns = ['MonthlyEarnings','YearlyEarnings'],inplace = True)
```

### From the above we can that all the columns are of type Object i.e. String except uploads, subscribers and videoviews. Min and Max Montly and Yearly earnings must be of Type Float but it is String currently. Also the column usercreated is Data but it is represented as String currently. So, we need to explicitly change the datatype of these columns

```python
# Changing the datatype to Float
df_youtube['MinMonthlyEarnings'] = df_youtube['MinMonthlyEarnings'].astype('float64')
df_youtube['MaxMonthlyEarnings'] = df_youtube['MaxMonthlyEarnings'].astype('float64')
df_youtube['MinYearlyEarnings'] = df_youtube['MinYearlyEarnings'].astype('float64')
df_youtube['MaxYearlyEarnings'] = df_youtube['MaxYearlyEarnings'].astype('float64')
```

### if we took a closer look into the data, the usercreated date is represented as jan 1st,2009 or Jan 3rd,2008. Next to date we have st,rd,th,nd which is not necessary. So, we can remove that by using replace method and specifing the regex expression

```python
# Removing st,nd,rd,th from the date
df_youtube['usercreated'] = df_youtube['usercreated'].str.replace(r'(st|nd|rd|th)', '', regex=True)
```

### We have specified the format as '%b %d %Y' %b - Month represented using first 3 alphabbets, %d date, %Y - Year which is the input date format in the dataset. It will convert this yyyy-mm-dd formt. Also some rows has invalid date format. some dates are represented as '--' which will throw an error when converting to datetime format. Inorder to ignore this we specified the parameter errors='coerce' which will ignore such errors and change the invalid date as NaT - not a valid Time representation which we can then replace with some default value.

```python
# Converting the usercreated column to datetime
df_youtube['usercreated'] = pd.to_datetime(df_youtube['usercreated'], format='%b %d, %Y', errors='coerce')
```

```python
# Filling the invalid date formats with a default value 2000-01-01
df_youtube['usercreated'] = df_youtube['usercreated'].fillna(pd.Timestamp('2000-01-01'))
```

```python
# Displaying the dataset after converting the usercreated as datetime
df_youtube
```

### We have created two different functions to find the distribution of channeltype. The first method returns a datframe which displays the distribution count and percentage of channeltype for the extracted top 1000 records. The second function visualized the distribution of channelType using a barchart. Both the functions take the dataframe as input parameter. we have used value_counts() method to get the count of exh channel type

```python
# Function to calculate the distribution count and percentage of channel type
def calDistributionChannelType(top_1000_records):
    # Count of each channeltype
    distribution = top_1000_records['channeltype'].value_counts()
    # percentage of each channel type
    percentage = (distribution / distribution.sum()) * 100
    #Converting to a dataframe
    result = pd.DataFrame({'Count': distribution, 'Percentage': percentage})
    return result
    
```

```python
# Importing pyplot from matplotlib library
import matplotlib.pyplot as plt
# Function to visualize the distribution of channel type
def visualizeChannelType(top_1000_records):
    # Setting the plot width and height
    plt.figure(figsize=(20,6))
    # Count of each channel type
    distribution = top_1000_records['channeltype'].value_counts()
    # creating a bar chart with channel type of x-axis and count on y-axis
    plt.bar(distribution.index, distribution.values)
    # Setting the x and y-axis label 
    plt.xlabel('Channel Type')
    plt.ylabel('Count')
    # Setting the title
    plt.title('Distribution of Channel Type')
    # Displaying/showing the plot
    plt.show()
```

```python
# extracting the top 1000 records using head method
top_1000_records = df_youtube.head(1000)
```
